Data Preprocessing

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
import numpy as np

# Step1: Random Sampling impulation for high null columns 
def random_impute(col):
    non_null = df[col].dropna()
    return df[col].apply(lambda x:np.random.choice(non_null) if pd.isnull(x) else x)

high_null_cols = ['rbc', 'rc', 'wc']
for col in high_null_cols:
    df[col] = random_impute(col)

# Step 2 : Mean impulation for numberical columns with low null values 
mean_impute_cols = ['pot', 'sod']
for col in mean_impute_cols:
    df[col] = df[col].fillna(df[col].mean())

# Step 3: Mode imputation for object  columns with low null values 
mode_impute_cols = ['htn', 'appet', 'pe', 'ane', 'dm',  'cad', 'pcc', 'ba']
for col in mode_impute_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Step 4:  fill  remaining numerical columns with mean if they have missing values
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in  numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

# Step 5: Fill reamaining object columns with mode
object_cols = df.select_dtypes(include = 'object').columns
for col in object_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

Model Building

```python
features = ['age', 'bp', 'sg', 'al', 'su', 'hemo', 'pcv', 'pot']
df = df[features + ['classification']]
```

```python
X = df[features]
y = df['classification']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state= 0)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
kidney_model = LogisticRegression()
kidney_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
# Predictions
y_pred = kidney_model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# plot_confusion_matrix(logmodel, X_test, y_test, cmap=plt.cm.Blues) # Old code
ConfusionMatrixDisplay.from_estimator(kidney_model, X_test, y_test, cmap=plt.cm.Blues)
plt.title('Confusion matrix for Logistic Regression', y=1.1)
plt.show()
```
